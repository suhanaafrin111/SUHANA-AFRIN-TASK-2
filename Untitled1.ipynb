```python
def circle_area(radius):
  area = 3.14 * radius * radius
  return area

# example
r = float(input("enter radius: "))
print("area of circle =", circle_area(r))
```

```python
def largest(a, b, c):
   if a>= b and a>= c:
      return a
   elif b>= a and b>= c:
    return b
   else:
    return c

 #example
num1 = int(input("enter first number: "))
num2 = int(input("enter second number: "))
num3 = int(input("enter third number: "))

print("largest number is:", largest(num1, num2, num3))
```

```python
def is_prime(num):
    if num <- 1:
      return false

    for i in range(2, num):
        if num % i == 0:
           return False

    return True

 #example
n = int(input("enter a number: "))

if is_prime(n):
   print("prime number")
else:
   print("not a prime number")

```

```python
def fibonacci(n):
  if n <= 1:
     return n
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)
#example
num = int(input("enter a number: "))
print("fibonacci number is:", fibonacci(num))
```

```python
def reverse_string(text):
   if len(text) ==0:
      return text
   else:
      return reverse_string(text[1:]) + text[0]
# example
s = input("enter a string: ")

print("reversed string is:", reverse_string(s))
```
